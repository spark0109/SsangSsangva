```python
list = []

for item in starbucks_csv["address"]:
    data = item.split()
    if ( data[0] == "서울특별시"):
        for data1 in data:
            if data1.find("길") > 0:
                if data1[-1] != "구":
                    list.append(data1)
            elif data1.find("로") > 0:
                if data1[-1] != "구":
                    list.append(data1)
df = pd.DataFrame(list).value_counts()
df
```
